## Развлекаемся со статистикой, начнем, пожалуй, с КРОССОВЕРОВ

```python
import pandas as pd
import numpy as np
```

```python
fanfictionDF = pd.read_csv('fanfiction.csv', index_col = 0)
fanfictionDF = fanfictionDF.replace('NAN', np.nan)
fanfictionDF
```

```python
fandomsDF = fanfictionDF[['fandoms']]
pd.unique(fandomsDF.fandoms)
```

## Пейринги

```python
pairingsDF = fanfictionDF[['pairings']]
pairingsDF = pairingsDF.assign(pairings=pairingsDF['pairings'].str.split(', ')).explode('pairings')
pairingsDF = pairingsDF.dropna()
pairingsDF
```

```python
pairings_sortedDF = pairingsDF.assign(pairings=pairingsDF['pairings'].str.split('/'))
pairings_sortedDF = pairings_sortedDF.pairings.apply(lambda x: '/'.join(sorted(x)))
```

```python
counted_pairingsDF = pairings_sortedDF.value_counts().rename_axis('pairing').reset_index(name='counts')
counted_pairingsDF.head(20)
```

```python
fanfictionDF[fanfictionDF.pairings.str.contains('Гарри Поттер/Джинни Уизли/Хедвиг', na=False)] 
```

```python
counted_pairingsDF[counted_pairingsDF.pairing.str.contains('Гермиона Грейнджер', na=False)] 
```

```python
counted_pairingsDF.to_excel('pairings_count.xlsx')
```

## События

```python
eventsDF = fanfictionDF[['events']]
eventsDF = eventsDF.assign(events=eventsDF['events'].str.split(', ')).explode('events')
eventsDF = eventsDF.dropna()
eventsDF
```

```python
counted_eventsDF = eventsDF.value_counts().rename_axis('event').reset_index(name='counts')
counted_eventsDF.head(20)
```

```python
counted_eventsDF.to_excel('events_count.xlsx')
```
